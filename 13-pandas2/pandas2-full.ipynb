### Movies by duration

```python
# We have missing values in the duration column
movies['duration'].isna().sum()
```

```python
def extract_duration(duration): 
    if pd.isna(duration): 
        return 0
    return int(duration.split()[0])
    
# Apply a function to each element in a column
movies['duration'].apply(extract_duration)
```

```python
# or use lambda function
movies['duration_converted'] = movies['duration'].apply(lambda x: 0 if pd.isna(x) else int(x.split()[0]))
```

### Movies by genre

```python
movies['listed_in']
```

```python
movie_genre = list(movies['listed_in'])
movie_genre[:10]
```

```python
movie_genre_dict = {}
for g in movie_genre: 
    g_parts = g.split(', ')
    for gg in g_parts:
        if gg not in movie_genre_dict: 
            movie_genre_dict[gg] = 0
        movie_genre_dict[gg] += 1
movie_genre_dict
```

```python
movie_genre_df = pd.DataFrame.from_dict(movie_genre_dict, orient="index", columns=["count"])
movie_genre_df
```

```python
movie_genre_df = movie_genre_df.sort_values(by='count')
movie_genre_df
```

```python
movie_genre_df.plot.bar()
```

### Stacked bar plot of genre based on country
Two ways of iterating a dataframe:
* `df.iterrows()`: Returns an iterator of (index, Series) pairs
* `df.itertuples()`: Returns an iterator of namedtuples for each row

```python
for row in movies.iterrows(): 
    print(row)
    break
```

```python
for row in movies.itertuples(): 
    print(row)
    break
```

```python
# indices are genres
movie_genre_df.index
```

```python
# columns are countries
movies['country'].unique()[:10]
```

```python
countries = {}
for c in movies['country']: 
    if pd.isna(c): 
        continue
    country_list = c.split(', ')
    for cc in country_list: 
        if cc not in countries: 
            countries[cc] = 0
        countries[cc] += 1
country_series = pd.Series(countries).sort_values(ascending=False)
country_series[:9]
```

```python
top5_countries = country_series[:5].index
top5_countries
```

```python
movie_genre_by_country_df = pd.DataFrame(0, index=movie_genre_df.index, 
                                        columns=top5_countries
                                        )
movie_genre_by_country_df
```

```python
for row in movies.itertuples(): 
    if pd.isna(row.country): 
        continue
    if pd.isna(row.listed_in): 
        continue
    countries = row.country.split(', ')
    genres = row.listed_in.split(', ')
    
    for c in countries: 
        for g in genres: 
            cc = c
            if c in top5_countries: 
                movie_genre_by_country_df.at[g, cc] += 1
```

```python
movie_genre_by_country_df
```

```python
movie_genre_by_country_df.plot(kind="bar", stacked=True)
```

```python
# Bar plot by using percentages
movie_genre_by_country_per_df = movie_genre_by_country_df.div(movie_genre_by_country_df.sum(axis=1), axis=0) * 100
movie_genre_by_country_per_df
```

```python
movie_genre_by_country_per_df.plot(kind="bar", stacked=True)
```

### Number of movies released by countries by year

```python
release_years = df['release_year'].value_counts().sort_index()
release_years
```

```python
movie_release_year_by_country_df = pd.DataFrame(0, index=release_years.index, 
                                        columns=top5_countries)

for row in movies.itertuples(): 
    if pd.isna(row.country): 
        continue
    if pd.isna(row.release_year): 
        continue
    countries = row.country.split(', ')
    
    for c in countries: 
        cc = c
        if c in top5_countries: 
            movie_release_year_by_country_df.at[row.release_year, cc] += 1
```
